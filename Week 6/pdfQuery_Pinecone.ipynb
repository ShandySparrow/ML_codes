```python
## Import libraries

import openai
import langchain
import pinecone
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import Pinecone
from langchain.llms import OpenAI 
```

```python
## lets get the pdf
def read_doc(directory):
    file_loader = PyPDFLoader(directory)
    documents = file_loader.load()
    return documents
```

```python
doc = read_doc("/Users/santhoshrajesh/Desktop/ML_road_map.pdf")
doc
```

```python
len(doc)
```

```python
## Divide Docs into chunks

def chunk_data(docs, chunk_size = 800, chunk_overlap = 50):
    text_splitter = RecursiveCharacterTextSplitter(chunk_size = chunk_size, chunk_overlap = chunk_overlap)
    doc = text_splitter.split_documents(docs)
    return doc
```

```python
documents = chunk_data(doc)
documents
```

```python
len(documents)
```

```python
## Embedding Technique of OpenAI

embeddings = OpenAIEmbeddings(api_key=os.environ['OPENAI_API_KEY'])
embeddings
```

```python
vectors = embeddings.embed_query("My name is shandy")
vectors
```

```python
len(vectors)
```

```python
pc = Pinecone.from_existing_index(
    index_name = "pdfquery",
    embedding = embeddings, 
    namespace = "shandy"
)
```

```python
index = pc.from_documents(doc, embeddings, index_name = "pdfquery")
```

```python
## Cosine Similarity Retrive Results

def retrieve_query(query, k=2):
    matching_results = index.similarity_search(query, k=k)
    return matching_results
```

```python
from langchain.chains.question_answering import load_qa_chain

llm = OpenAI(model_name = "gpt-3.5-turbo-instruct", temperature = 0.5)
chain = load_qa_chain(llm, chain_type="stuff")
```

```python
## Search answers from vectorDB

def retrieve_answers(query):
    doc_search = retrieve_query(query)
    print(doc_search)
    response = chain.run(input_documents = doc_search, question = query)
    
    return response
```

```python
our_query = "What are the concepts said to learn under cloud computing ?"
answer = retrieve_answers(our_query)
print(answer)
```
