```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd
```

```python
# Example text data (list of documents)
text_data = [
    "text mining is interesting",
    "association rules are powerful",
    "mining techniques for text data",
    "discover patterns in text"
]
```

```python
# Text preprocessing and representation
te = TransactionEncoder()
te_ary = te.fit(text_data).transform(text_data)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
df
```

```python
df.columns
```

```python
# Apriori algorithm
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)
frequent_itemsets
```

```python
# Association rules
rules = association_rules(frequent_itemsets.iloc[:5000,:], metric='confidence', min_threshold=0.7)

# Print the rules
rules
```
